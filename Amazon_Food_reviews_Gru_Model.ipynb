```python
train_df_y = to_categorical(train_df_y)
```

```python
max_num_words =10000 # hp that select unique 10000 words
seq_len = 50 # hp from each document it select 50 words out of 10000
embedding_size =100 # hp vector length of each word

model = Sequential()
model.add(Embedding(max_num_words, embedding_size, input_length = seq_len))
model.add(GRU(100, return_sequences ='true'))
model.add(GRU(100, return_sequences ='true'))
model.add(GRU(64))
model.add(Dense(2, activation ='softmax'))

adam=Adam(learning_rate=0.01)
model.compile(optimizer=adam,loss='categorical_crossentropy', metrics=['accuracy'])

tokenizer = Tokenizer(num_words= max_num_words)
tokenizer.fit_on_texts(df.X)

train_df_x = tokenizer.texts_to_sequences(train_df_x) # will convert the text to sequences of Ids
train_df_x = pad_sequences(train_df_x, maxlen= seq_len)
test_df_x = tokenizer.texts_to_sequences(test_df_x) # will convert the text to sequences of Ids
test_df_x = pad_sequences(test_df_x, maxlen= seq_len)

```

```python
model.fit(train_df_x, train_df_y, validation_split=0.2, epochs=2,batch_size=256)
```
